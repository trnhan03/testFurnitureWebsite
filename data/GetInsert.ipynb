```python
Data = pd.ExcelFile('Data.xlsx')
Table = Data.sheet_names
```

```python
def Get_print_value(Data_in, Name):
    It = Data_in.parse(Name)
    print(f"INSERT INTO {Name}")
    print(f"VALUE")
    temp = ""
    for id, row in It.iterrows():
        temp += f"\t("
        for item in row:
            temp += f"'{item}',"
        temp = temp[:-1]
        temp += f"),\n"
    temp = temp[:-2]
    temp += f";\n"
    print(temp)
```

```python
for item in Table:
    Get_print_value(Data, item)
```

```python
import os
import shutil

df = Data.parse("PRODUCT")
df = df[["ID", "Name"]]
df
```

```python
for id, row in df.iterrows(): 
    current_folder_name = row[1]
    new_folder_name = row[0]
    
    # Check if the current folder exists
    if os.path.exists(current_folder_name):
        try:
            # Rename the folder
            os.rename(current_folder_name, new_folder_name)
            print(f'Renamed folder: {current_folder_name} to {new_folder_name}')
        except Exception as e:
            print(f'Error renaming folder: {str(e)}')
    else:
        print(f'Folder {current_folder_name} does not exist')
```
